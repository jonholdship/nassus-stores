# Item Tables
The purpose of this notebook is to explore our data tables and produce tables of appearance probabilities and mark ups.

```python
import pandas as pd
from os.path import join
import numpy as np
data_folder="../static/item-data"
```

```python
general_store_df=pd.read_csv(join(data_folder,"mundane.csv"))
general_store_df=general_store_df[["Type","Subtype"]].drop_duplicates()
general_store_df["probability"]=0.0
general_store_df["mark_up"]=1.2
general_store_df["Level"]=1
```

```python
general_store_df.loc[general_store_df["Type"]=="Misc","probability"]=0.8
```

```python
general_store_df.loc[general_store_df["Subtype"]=="Misc","probability"]=0.7
general_store_df.loc[general_store_df["Subtype"]=="Light","probability"]=0.2
general_store_df.loc[general_store_df["Subtype"]=="Simple","probability"]=0.2
```

```python
general_store_df.to_csv(join(data_folder,"general_store.csv"))
```
